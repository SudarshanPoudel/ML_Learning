```python
#Import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

Plotting the data

```python
#Sine wave
plt.plot(x, y)
plt.show()
```

```python
#Build cosine wave
plt.plot(x, z)
plt.show()
```

```python
#Adding title, X axis, y axix labels
plt.plot(x, y)
plt.xlabel('Angle')
plt.ylabel('Sine values')
plt.title('Sine wave')
plt.show()
```

```python
#Parabola
x = np.linspace(-10, 10, 20)
y = x**2
plt.plot(x, y)
plt.show()
```

```python
#Plotting costume graphs
plt.plot(x, y, 'b+') # r = red, b = blue etc. ^, ., * ... can be used as symbol
plt.show()
```

```python
#Multiple graph in single graph
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'b--')
plt.show()
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
languages = ['English', 'French', 'Spanish', 'Nepali', 'Hindi']
people = [100, 180, 55, 48, 88]
ax.bar(languages, people)
plt.xlabel('Languages')
plt.ylabel('No of people')
plt.show()
```

Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0, 0, 1, 1])
languages = ['English', 'French', 'Spanish', 'Nepali', 'Hindi']
people = [100, 180, 55, 48, 88]
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

```python
x = np.linspace(0,100, 50)
y = np.random.randint(0, 50, 50)
z = np.random.randint(40, 100, 50)
fig2 = plt.figure()
ax = fig2.add_axes([0, 0, 1, 1])
ax.scatter(x, y, color='g')
ax.scatter(x, z, color='b')
plt.show()
```

3d scatter diagram

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20* np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x, y, z, c = z, cmap = 'Blues')
plt.show()
```
