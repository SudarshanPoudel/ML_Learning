```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Note : Seaborn has some build in datasets

```python
#Total bill vs tip dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
#Setting defaulyt theme for theme
sns.set_theme()
```

```python
#visualize the data
sns.relplot(data=tips, x = 'total_bill', y = 'tip', col = 'time', hue = 'smoker', style = 'smoker', size = 'size')
```

Scatter plot

```python
#House price predction
from sklearn.datasets import fetch_california_housing
california_dataset = fetch_california_housing()  
house = pd.DataFrame(california_dataset.data, columns=california_dataset.feature_names)
house['Price'] = california_dataset.target

house.head()
```

Distribution Plot

```python
sns.distplot(house['Price'])
```

Correlation

```python
correlation = house.corr()
```

```python
#Constructing a heatmap
plt.figure(figsize=(5, 5))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size': 8}, cmap='Blues')
```
